Dummy Variables

Preprocessing few discrete variables

```python
pd.get_dummies(loan_data_test['grade'])
```

```python
pd.get_dummies(loan_data_test['grade'],prefix='grade', prefix_sep=' : ').astype(int).head()
```

```python
loan_data_dummies = [pd.get_dummies(loan_data_test['grade'], prefix='grade', prefix_sep=' : '),
                     pd.get_dummies(loan_data_test['sub_grade'], prefix='sub_grade', prefix_sep=' : '),
                     pd.get_dummies(loan_data_test['home_ownership'], prefix='home_ownership', prefix_sep=' : '),
                     pd.get_dummies(loan_data_test['verification_status'], prefix='verification_status', prefix_sep=' : '),
                     pd.get_dummies(loan_data_test['loan_status'], prefix='loan_status', prefix_sep=' : '),
                     pd.get_dummies(loan_data_test['purpose'], prefix='purpose', prefix_sep=' : '),
                     pd.get_dummies(loan_data_test['addr_state'], prefix='addr_state', prefix_sep=' : '),
                     pd.get_dummies(loan_data_test['initial_list_status'], prefix='initial_list_status', prefix_sep=' : ')
                     
                     
                     ]
```

```python
loan_data_dummies = pd.concat(loan_data_dummies, axis = 1 )
```

```python
type(loan_data_dummies)
```

```python
loan_data_test = pd.concat([loan_data_test, loan_data_dummies], axis = 1)
```
